```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.common.by import By
import time
from PIL import Image
import base64
from io import BytesIO
import tkinter as tk
from tkinter import simpledialog
import os
```

```python

# Function to display captcha image and get user input
def get_captcha_input(image_base64):
    # Decode the base64 image
    img_data = base64.b64decode(image_base64.split(',')[1])
    img = Image.open(BytesIO(img_data))
    
    # Save the captcha image locally
    img.save("captcha.png")
    
    # Display the image using Tkinter
    root = tk.Tk()
    root.title("Captcha")
    img = tk.PhotoImage(file="captcha.png")
    panel = tk.Label(root, image=img)
    panel.pack(side="top", fill="both", expand="yes")
    
    # Prompt the user to enter the captcha
    captcha_value = simpledialog.askstring("Input", "Please enter the captcha text:")
    
    root.destroy()  # Close the Tkinter window
    os.remove("captcha.png")  # Clean up the captcha image
    return captcha_value
```

```python
from selenium.webdriver.firefox.options import Options

# Set up the WebDriver (assuming you are using Firefox)
service = Service('/usr/bin/geckodriver')  # Path to the GeckoDriver
options = Options()
options.binary_location = '/usr/bin/firefox'  # Path to the Firefox binary

driver = webdriver.Firefox(service=service, options=options)

# Open the login page
driver.get('https://tms20.nepsetms.com.np/login')

# Wait for the page to load
time.sleep(2)

# Locate and fill the username field
username = driver.find_element(By.XPATH, '//input[@placeholder="Client Code/ User Name"]')
username.send_keys('your_username')

# Locate and fill the password field
password = driver.find_element(By.XPATH, '//input[@placeholder="Password"]')
password.send_keys('your_password')

# Locate the captcha image element
captcha_image_element = driver.find_element(By.XPATH, '//img[@class="form-control captcha-image-dimension col-10"]')

# Extract the captcha image as a base64 encoded string
captcha_image_base64 = driver.execute_script("""
    var canvas = document.createElement('canvas');
    var context = canvas.getContext('2d');
    var img = arguments[0];
    canvas.width = img.width;
    canvas.height = img.height;
    context.drawImage(img, 0, 0, img.width, img.height);
    return canvas.toDataURL();
    """, captcha_image_element)

# Display captcha and get the input from the user
captcha_value = get_captcha_input(captcha_image_base64)

# Locate and fill the captcha field
captcha = driver.find_element(By.XPATH, '//input[@placeholder="Enter Captcha"]')
captcha.send_keys(captcha_value)

# Submit the form
submit_button = driver.find_element(By.XPATH, '//input[@value="Login"]')
submit_button.click()

# Wait for a while to observe the result (optional)
time.sleep(5)

# Close the browser
driver.quit()
```
